```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(0)
num_samples = 1000
colors = ['Red', 'Blue', 'Green', 'Crimson', 'yellow']
nominal_data = np.random.choice(colors, num_samples)
nominal_data
```

```python
dataframe = pd.DataFrame(nominal_data, columns=['Colors'])
dataframe
```

Data creation of the ordinal column named 'satisfaction_levels' satisfacion_levels = ['Low', 'Medium', 'High']

```python
np.random.seed(0)
satisfaction_level = ['Low', 'Medium', 'High']
ordinal_data = np.random.choice(satisfaction_level, num_samples)
dataframe['Satisfaction Levels'] = ordinal_data
dataframe
```

Data creatation of the interval data under temperature column

```python
np.random.seed(0)
interval_data = np.random.randint(0, 100, num_samples)
dataframe['temperature'] = interval_data
dataframe
```

Data creation of the range data under the income column

```python
np.random.seed(0)
range_data = np.random.randint(10000, 200000, num_samples)
dataframe['income'] = range_data
dataframe
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(15, 10))
plt.hist(dataframe['income'], bins=20, color='green', edgecolor='black')
plt.title('Income Distribution')
plt.ylabel('Frequency')
plt.xlabel('Income (Rupees)')
plt.show()
```
