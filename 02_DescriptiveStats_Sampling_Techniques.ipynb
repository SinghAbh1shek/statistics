```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/Population_Survey_Data_lyst1750274390196.csv')
data.head()
```

```python
data.shape
```

1. Simple random sampling: 100 records randomly

```python
sample_data = data.sample(n=100, random_state = 44)
sample_data
```

2. Stratified Random Sampling

```python
Stratified_sampling = data.groupby('Region').apply(lambda x: x.sample(min(len(x), 25)))
Stratified_sampling
```

Cluster Sampling: Manufactring data by batch number

```python
manufacturing_data = pd.read_csv("/content/Manufacturing_Data_lyst1750274395200.csv")
manufacturing_data.head()
```

```python
clusters = manufacturing_data['BatchNumber'].unique()
clusters
```

```python
selected_clusters = np.random.choice(clusters, size=5, replace=False, )
selected_clusters
```

Systemetic Sampling

```python
clusetred_sample = manufacturing_data[manufacturing_data['BatchNumber'].isin(selected_clusters)]
clusetred_sample
```

```python
systematic_sample = data.iloc[::10, :]
systematic_sample
```

Measures of central tendency: Mean, Median, Mode

```python
data.head()
```

Q. Evaluate the measure of the central tendency for the whole data follwed by simple random sample, Stratified Random Sample, Cluster Sample, and systematic sample

Compare the mean of all the different types of sample
Finally, conclude which spacific sampling technique mean value is the almost similar to the population mean value.

```python
data_mean = data['Income'].mean()
print("Data mean: ", data_mean)
manufacturing_mean = manufacturing_data['BatchNumber'].mean()
print("ManufacturingData Mean: ", manufacturing_mean)

```

```python
simple_random_mean = np.mean(sample_data['Income'])
print("Simple Random Sample Mean: ", simple_random_mean)


Stratified_mean = np.mean(Stratified_sampling['Income'])
print("Stratified Random Sample Mean: ", Stratified_mean)


clusetred_mean = np.mean(clusetred_sample['BatchNumber'])
print("Clustered Sample Mean: ", clusetred_mean)


systematic_sample_mean = np.mean(systematic_sample['Income'])
print("Systematic Sample Mean: ", systematic_sample_mean)
```

```python
data_median = np.median(data['Income'])
print("Data median: ", data_median)
manufacturing_median = np.median(manufacturing_data['BatchNumber'])
print("ManufacturingData median: ", manufacturing_median)
```

```python
simple_random_median = np.median(sample_data['Income'])
print("Simple Random Sample median: ", simple_random_median)


Stratified_median = np.median(Stratified_sampling['Income'])
print("Stratified Random Sample median: ", Stratified_median)


clusetred_median = np.median(clusetred_sample['BatchNumber'])
print("Clustered Sample median: ", clusetred_median)


systematic_sample_median = np.median(systematic_sample['Income'])
print("Systematic Sample median: ", systematic_sample_median)
```
