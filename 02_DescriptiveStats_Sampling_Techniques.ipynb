```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/Population_Survey_Data_lyst1750274390196.csv')
data.head()
```

```python
data.shape
```

1. Simple random sampling: 100 records randomly

```python
sample_data = data.sample(n=100, random_state = 44)
sample_data
```

2. Stratified Random Sampling

```python
Stratified_sampling = data.groupby('Region').apply(lambda x: x.sample(min(len(x), 25)))
Stratified_sampling
```

Cluster Sampling: Manufactring data by batch number

```python
manufacturing_data = pd.read_csv("/content/Manufacturing_Data_lyst1750274395200.csv")
manufacturing_data.head()
```

```python
clusters = manufacturing_data['BatchNumber'].unique()
clusters
```

```python
selected_clusters = np.random.choice(clusters, size=5, replace=False, )
selected_clusters
```

Systemetic Sampling

```python
clusetred_sample = manufacturing_data[manufacturing_data['BatchNumber'].isin(selected_clusters)]
clusetred_sample
```

```python
population_cluster = data['Age'].unique()
population_cluster
```

```python
selected_clusters1 = np.random.choice(population_cluster, size=5, replace=False, )
selected_clusters1
```

```python
clusetred_sample1 = data[data['Age'].isin(selected_clusters1)]
clusetred_sample1
```

```python
systematic_sample = data.iloc[::10, :]
systematic_sample
```

Measures of central tendency: Mean, Median, Mode

```python
data.head()
```

Q. Evaluate the measure of the central tendency for the whole data follwed by simple random sample, Stratified Random Sample, Cluster Sample, and systematic sample

Compare the mean of all the different types of sample
Finally, conclude which spacific sampling technique mean value is the almost similar to the population mean value.

```python
data_mean = data['Income'].mean()
data_median = data['Income'].median()
data_mode = data['Income'].mode()[0]


print("Data mean: ", data_mean)
print("Data median: ", data_median)
print("Data mode: ", data_mode)
```

```python
random_mean = sample_data['Income'].mean()
random_median = sample_data['Income'].median()
random_mode = sample_data['Income'].mode()[0]


print("Simple Random Sample Mean: ", random_mean)
print("Simple Random Sample Median: ", random_median)
print("Simple Random Sample Mode: ", random_mode)
```

```python
Stratified_mean = Stratified_sampling['Income'].mean()
Stratified_median = Stratified_sampling['Income'].median()
Stratified_mode = Stratified_sampling['Income'].mode()[0]


print("Stratified Randon Sample Mean: ", Stratified_mean)
print("Stratified Randon Sample Median: ", Stratified_median)
print("Stratified Randon Sample Mode: ", Stratified_mode)
```

```python
clusetred_mean = clusetred_sample1['Income'].mean()
clusetred_median = clusetred_sample1['Income'].median()
clusetred_mode = clusetred_sample1['Income'].mode()[0]

print("Clustred Sample Mean: ", clusetred_mean)
print("Clustred Sample Median: ", clusetred_median)
print("Clustred Sample Mode: ", clusetred_mode)
```

```python
systematic_mean = systematic_sample['Income'].mean()
systematic_median = systematic_sample['Income'].median()
systematic_mode = systematic_sample['Income'].mode()[0]


print("Systematic Sample Mean: ", systematic_mean)
print("Systematic Sample Median: ", systematic_median)
print("Systematic Sample Mode: ", systematic_mode)
```

```python
data_v = {
    "method": ["Random", "Stratified", "Cluster", "Systematic"],
    "mean": [random_mean, Stratified_mean, clusetred_mean, systematic_mean]
}
ax = pd.DataFrame(data_v).plot.bar(x="method", y="mean")
true_median = data_median
plt.axhline(y=true_median, color='red', linestyle='--')
```

```python
data_medn = {
    "method": ["Random", "Stratified", "Cluster", "Systematic"],
    "median": [random_median, Stratified_median, clusetred_median, systematic_median]
}
ax = pd.DataFrame(data_medn).plot.bar(x="method", y="median")
true_mean = data_median
plt.axhline(y=true_mean, color='red', linestyle='--')
```

```python
data_mod = {
    "method": ["Random", "Stratified", "Cluster", "Systematic"],
    "mode": [random_mode, Stratified_mode, clusetred_mode, systematic_mode]
}
ax = pd.DataFrame(data_mod).plot.bar(x="method", y="mode")
true_mean = data_mode
plt.axhline(y=true_mean, color='red', linestyle='--')
```
