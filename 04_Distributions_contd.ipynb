```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(42)

# Standard Normal Distribution
portfolio_returns = np.random.normal(loc=0, scale=1, size=1000)
```

```python
# Calculate the kurtosis


from scipy.stats import kurtosis
kurt_value = kurtosis(portfolio_returns)
print(f"Kurtosis: {kurt_value:.2f}")
```

```python
sns.histplot(portfolio_returns, bins=30, kde=True)
plt.title(f"Portfolio Return Distributions (Kurtosis: {kurt_value:.2f})")
plt.xlabel("Returns")
plt.ylabel("Frequency")
plt.show()
```

```python
np.random.seed(42)

portfolio_returns = np.random.normal(loc=0, scale=1, size=1000)

# Adding the extremes value
portfolio_returns = np.random.normal(loc=0, scale=50, size=50)

# Calculate the kurtosis
from scipy.stats import kurtosis
kurt_value = kurtosis(portfolio_returns)
print(f"Kurtosis: {kurt_value:.2f}")

sns.histplot(portfolio_returns, bins=30, kde=True)
plt.title(f"Portfolio Return Distributions (Kurtosis: {kurt_value:.2f})")
plt.xlabel("Returns")
plt.ylabel("Frequency")
plt.show()
```
