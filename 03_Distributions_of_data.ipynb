```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import skew
```

```python
# Data creations for the income

np.random.seed(42)

# Right skewed data
income_distributions = np.random.gamma(shape=2.0, scale=1000.0, size=1000)

# Calcullate the skewness
skew_value = skew(income_distributions)
```

```python
sns.histplot(income_distributions, bins=30, kde=True)
plt.title(f"Income Distribution (Skewness: {skew_value:.2f})")
plt.xlabel("Income")
plt.ylabel("Frequency")
```
